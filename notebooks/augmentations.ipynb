```python
def augmentation_experiments(base_config):
    experiments = []

    for single_config in base_config:
        for aug in [False, True]:
            for mix in [None, "mixup", "cutmix", "both"]:

                config = single_config.copy()

                config["augmentation"] = aug
                config["mix_type"] = mix


                experiments.append(config)

    return experiments
```

```python
#exaple of best params found by random search
best_params = {
    "lr": 5e-4,
    "weight_decay": 1e-4,
    "dropout": 0.3,
    "batch_size": 128,
}
```

```python
search_space = {
    "model": ["smallcnn", "bigcnn"],

    "lr": [best_params["lr"]],
    "weight_decay": [best_params["weight_decay"]],
    "dropout": [best_params["dropout"]],
    "batch_size": [best_params["batch_size"]],

    "augmentation": [True],

    "mix_type": [None, "mixup", "cutmix", "both"],
    "alpha": [0.2, 0.5, 1.0],
    "mix_prob": [0.5, 1.0],
}
```

```python
run_all_experiments(
    search_space,
    n_samples=15,
    folder="exp2_aug"
)
```
